###3. Pengecekan Missing Value pada Dataset

Tujuan pengecekan nilai hilang yaitu untuk menghindari error pada saat analisis uji pengaruh menggunakan uji regresi berganda. Selain itu, jumlah nilai hilang yang ada pada suatu dataset bisa memengaruhi kualitas dataset yang dapat menyebabkan kualitas pada hasil penelitian.

```python
import numpy as np

dates = df['Tahun'].copy()
df = df.drop(columns=['Tahun'])
missing_summary = []

for col in df.columns:
    total_missing = df[col].apply(lambda x: pd.isna(x)).sum()
    total_data = len(df[col])

    missing_summary.append({
        'Variabel': col,
        'Jumlah Missing Value': total_missing,
        'Jumlah Data': total_data
    })

summary_df = pd.DataFrame(missing_summary)
print("\n Ringkasan Missing Value Data Mentah per Variabel:")
print(summary_df.to_string(index=False))
```

Berdasarkan hasil pengecekan tersebut, terlihat bahwa dataset tidak memiliki nilai hilang, artinya tidak perlu dilakukan penanganan nilai hilang. Sehingga dataset dapat dilanjutkan ke analisis selanjutnya.

###4. Pengecekan Distribusi Statistik Data

Pengecekan distribusi data dengan beberapa pengukuran statistik bertujuan untuk mengetahui persebaran data yang digunakan, karakteristik lapangan berdasarkan data.

```python
import numpy as np

# Daftar variabel fitur dan target
fitur_variabel = [
"Rata-Rata Suhu", "Kelembapan Udara", "Jumlah Curah Hujan", "Lama Penyinaran Matahari",
"Kecepatan Angin Rata-Rata",
]
target_variabel = "Jumlah Kunjungan Wisatawan"

# Pengecekan nilai NaN pada setip atribut/variabel
df_clean = df[fitur_variabel + [target_variabel]].dropna()

# Analisis Deskriptif Statistik
stat_summary = []
for col in fitur_variabel + [target_variabel]:
    if np.issubdtype(df_clean[col].dtype, np.number):
        summary = {
            'Variabel': col,
            'Mean': df_clean[col].mean(),
            'Median': df_clean[col].median(),
            'Modus': df_clean[col].mode().iloc[0] if not df_clean[col].mode().empty else 'N/A',
            'Varian': df_clean[col].var(),
            'Standar Deviasi': df_clean[col].std()
        }
    else:
        summary = {
            'Variabel': col,
            'Mean': 'N/A',
            'Median': 'N/A',
            'Modus': df_clean[col].mode().iloc[0] if not df_clean[col].mode().empty else 'N/A',
            'Varian': 'N/A',
            'Standar Deviasi': 'N/A'
        }
    stat_summary.append(summary)

# Output Tabel Analisis Deskriptif Statistik
deskriptif_df = pd.DataFrame(stat_summary)
print("\n=== Tabel Analisis Deskriptif Statistik ===")
print(deskriptif_df.to_string(index=False))
```

Berdasarkan hasil pengukuran statistik diatas, dapat disimpulkan bahwa untuk nilai rata-rata suhu di Kota Semarang yaitu 28 derajat celcius, lalu untuk kelembapan udara cenderung tinggi yaitu sebesar 78%, jumlah curah hujan di Kota Semarang memiliki rata-rata sebesar 53 mm, lama penyinaran matahari di Kota Semarang selama 6 jam, dan untuk kecepatan angin rata-rata sebesar 2 m/s. Serta untuk jumlah pengunjung rata-rata di Kota Semarang untuk berwisata dari tahun 2019-2024 sebanyak 695469 orang.

###5. Normalisasi Data

Normalisasi dilakukan dengan tujuan untuk menghilangkan gap atau ketimpangan skala nilai kelas yang besar antara nilai kelas yang rendah dengan nilai kelas yang tinggi. Normalisasi data menghasilkan distribusi data memiliki nilai rata-rata mendekati 0 dan standari deviasi atau simpangan baku mendekati 1. Pada penelitian ini, normalisasi data dilakukan dengan menggunakan metode Min Max, dengan melakukan selisih dan pembagian menggunakan data x_i, lalu nilai minimum pada variabel tersebut, dan nilai maksimum pada variabel tersebut.

```python
import numpy as np

numeric_to_scale = [
   "Rata-Rata Suhu", "Kelembapan Udara", "Jumlah Curah Hujan", "Lama Penyinaran Matahari",
   "Kecepatan Angin Rata-Rata"
]

min_vals = df[numeric_to_scale].min()
max_vals = df[numeric_to_scale].max()
range_vals = (max_vals - min_vals).replace(0, 1)

minmax_scaled = (df[numeric_to_scale] - min_vals) / range_vals

X_scaled = df.copy()
X_scaled[numeric_to_scale] = minmax_scaled
X_scaled["Jumlah Kunjungan Wisatawan"] = df["Jumlah Kunjungan Wisatawan"]

unique_summary = []
for i, col in enumerate(numeric_to_scale + ["Jumlah Kunjungan Wisatawan"]):
    uniq_vals = np.unique(X_scaled[col])
    unique_summary.append({
        'No': i + 1,
        'Variabel': col,
        'Nilai Unik': uniq_vals[:10],
        'Jumlah Unik': len(uniq_vals)
    })
unique_df = pd.DataFrame(unique_summary)

print("\n=== Nilai Unik Setelah Normalisasi (kolom terpilih) ===")
print(unique_df.to_string(index=False))

top10_scaled = X_scaled[numeric_to_scale + ["Jumlah Kunjungan Wisatawan"]].copy()
top10_scaled.insert(0, "No", range(1, len(top10_scaled) + 1))
print("\nData Hasil Normalisasi Min-Max (seluruh data pada kolom terpilih)")
print(top10_scaled.to_string(index=False))

result_df = X_scaled.copy()

with pd.ExcelWriter("Dataset_Hasil_Normalisasi_MinMax.xlsx") as writer:
    result_df.to_excel(writer, sheet_name="Data Normalisasi", index=False)
    unique_df.to_excel(writer, sheet_name="Ringkasan Nilai Unik", index=False)
```

###6. Uji Regresi Linear Berganda

Tujuan dari uji regresi ini yaitu untuk mengetahui apakah variabel cuaca memiliki pengaruh secara signifikan terhadap variabel jumlah kunjungan wisatawan di Kota Semarang. Analisis hanya dilihat pada beberapa parameter, seperti nilai R-Squared yang menunjukan persentase pengaruh secara keseluruhan antara variabel independen dengan variabel dependen, lalu parameter P-Value yang merupakan nilai korelasi antara variabel independen terhadap variabel dependen. Jika nilai P-Value bernilai < 0.05, maka variabel tersebut memiliki pengaruh langsung secara signifikan terhadap variabel dependen dengan error sebesar 5%. Selain uji regresi, pada bagian ini akan ditampilkan hasil perhitungan dari pengukuran statistik seperti mean, median, modus, varian, dan standar deviasi.

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

df = pd.read_excel('Dataset_Hasil_Normalisasi_MinMax.xlsx')
# Daftar variabel fitur dan target
fitur_variabel = [
"Rata-Rata Suhu", "Kelembapan Udara", "Jumlah Curah Hujan", "Lama Penyinaran Matahari",
"Kecepatan Angin Rata-Rata",
]
target_variabel = "Jumlah Kunjungan Wisatawan"

# Pengecekan nilai NaN pada setip atribut/variabel
df_clean = df[fitur_variabel + [target_variabel]].dropna()

# Analisis Deskriptif Statistik
stat_summary = []
for col in fitur_variabel + [target_variabel]:
    if np.issubdtype(df_clean[col].dtype, np.number):
        summary = {
            'Variabel': col,
            'Mean': df_clean[col].mean(),
            'Median': df_clean[col].median(),
            'Modus': df_clean[col].mode().iloc[0] if not df_clean[col].mode().empty else 'N/A',
            'Varian': df_clean[col].var(),
            'Standar Deviasi': df_clean[col].std()
        }
    else:
        summary = {
            'Variabel': col,
            'Mean': 'N/A',
            'Median': 'N/A',
            'Modus': df_clean[col].mode().iloc[0] if not df_clean[col].mode().empty else 'N/A',
            'Varian': 'N/A',
            'Standar Deviasi': 'N/A'
        }
    stat_summary.append(summary)

# Output Tabel Analisis Deskriptif Statistik
deskriptif_df = pd.DataFrame(stat_summary)
print("\n=== Tabel Analisis Deskriptif Statistik ===")
print(deskriptif_df.to_string(index=False))

# Definisikan X (fitur) dan y (target)
X = df_clean[fitur_variabel]
y = df_clean[target_variabel]

# Ubah variabel kategorikal menjadi dummy
X = pd.get_dummies(X, drop_first=True)

# Tambahkan konstanta untuk intercept (b0)
X = sm.add_constant(X)

# Inisialisasi dan fit model OLS
model = sm.OLS(y, X).fit()

# Tampilkan hasil
print("\n=== Hasil Analisis Regresi dan Korelasi ===")
print(model.summary())
```

Berdasarkan pengukuran statistik pada data yang telah dilakukan normalisasi, untuk nilai per variabelnya rata-rata untuk nilai mean sudah mendekati 0 dan standar deviasi mendekati 1. Artinya proses normalisasi yang dilakukan sudah berhasil.

Lalu, untuk hasil dari uji regresi, seluruh variabel memiliki pengaruh secara signifikan terhadap variabel target (berdasarkan nilai Probabilitas F-Statistic) sebesar 32% (berdasarkan nilai R-Squared), artinya variabel cuaca hanya memiliki pengaruh terhadap jumlah kunjungan wisatawan di Kota Semarang sebanyak 32%, dan masih terdapat 68% pengaruh lainnya yang tidak diketahui. Berdasarkan hal ini, artinya variabel cuaca hanya memiliki pengaruh yang kecil terhadap jumlah fluktuasi wisatawan yang berkunjung ke Kota Semarang.

Selain itu, jika dilihat berdasarkan korelasi pada masing-masing variabel independen terhadap dependen, variabel independen yang memiliki pengaruh secara signifikan hanya variabel lama penyinaran matahari, sedangkan variabel cuaca lainnya tidak memiliki pengaruh secara signifikan terhadap kenaikan jumlah wisatawan di Kota Semarang.

###7. Grafik Persebaran Variabel Independen Terhadap Variabel Dependen

Disini akan disajikan visualisasi hasil terkait persebaran pengaruh antara variabel independen terhadap variabel dependen melalui grafik scatter plot.

```python
import matplotlib.pyplot as plt
import seaborn as sns

y_pred = model.predict(X)

plt.figure(figsize=(8,6))
sns.scatterplot(x=y, y=y_pred, alpha=0.7)
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linestyle='--')  # garis 45 derajat
plt.xlabel("Jumlah Kunjungan Wisatawan (Aktual)")
plt.ylabel("Jumlah Kunjungan Wisatawan (Prediksi)")
plt.title("Scatter Plot hasil prediksi jumlah wisatawan Kota Semarang")
plt.grid(True)
plt.show()

residuals = y - y_pred

plt.figure(figsize=(8,6))
sns.scatterplot(x=y_pred, y=residuals, alpha=0.7)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Jumlah Kunjungan Wisatawan (Prediksi)")
plt.ylabel("Residual")
plt.title("Scatter Plot Hasil residual analisis terhadap hasil Prediksi")
plt.grid(True)
plt.show()
```

Berdasarkan scatter plot diatas, jika akan melakukan prediksi jumlah kunjungan wisatawan di Kota Semarang hanya berdasarkan variabel cuaca, hasilnya akan memiliki akurasi yang rendah. Hal ini dikarenakan variabel cuaca tidak memiliki pengaruh yang besar terdapat jumlah kunjungan wisatawan yaitu hanya sebesar 32%, hal ini menyebabkan hasil prediksi masih kurang tepat.
